```python
faces_cascade=cv2.CascadeClassifier(r"D:\vs code\image classification\haarcascade_frontalface_default.xml") # to detect faces in the image
faces_cascade

```

```python
imgray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) # to convert the image to grey scale
plt.imshow(imgray,cmap='grey') # to display the image in grey scale
faces=faces_cascade.detectMultiScale(imgray,scaleFactor=1.1,minNeighbors=5) # to detect faces in the image
```

```python
faces
# it returns the coordinates of the faces detected in the image
```

```python
len(faces) # to get the number of faces detected in the image
```

```python
for (x,y,w,h) in faces: # to draw a rectangle around the faces detected in the image
    cv2.rectangle(converted_img,(x,y),(x+w,y+h),(0,255,0),2) # (img,pt1,pt2,color,thickness)
    plt.imshow(converted_img)
```

```python
vid=cv2.VideoCapture(r"D:\vs code\image classification\103255-662114536.mp4") # to read the video file
vid
```

```python
while True: # to read the video file frame by frame
    success,frame=vid.read() # to read the video file frame by frame
    grey=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY) # to convert the image to grey scale
    faces=faces_cascade.detectMultiScale(grey,scaleFactor=1.1,minNeighbors=5) # to detect faces in the image
    for (x,y,w,h) in faces: # to draw a rectangle around the faces detected in the image
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),2) # (img,pt1,pt2,color,thickness)
    cv2.imshow("video",frame) # to display the video file
    if cv2.waitKey(1) & 0xFF==ord('q'): # to exit the video file and waitkey 1 is used for the delay of 1millisecond
        #if the key 'q' is pressed, then exit the video file
        break
vid.release() # to release the video file
cv2.destroyAllWindows() # to destroy all the windows
```

```python
# to detect the eyes in the image
eye_cascade=cv2.CascadeClassifier(r"D:\vs code\image classification\haarcascade_eye.xml") # to detect eyes in the image
eye_cascade
```

```python
vid=cv2.VideoCapture(r"D:\vs code\image classification\103255-662114536.mp4") # to read the video file
vid
```

```python
vid=cv2.VideoCapture(0)
while True:
    success,frame=vid.read()
    grey=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    eye=eye_cascade.detectMultiScale(grey,scaleFactor=1.1,minNeighbors=5)
    for (x,y,w,h) in eye:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),2)
    cv2.imshow("eye_detection",frame)
    if cv2.waitKey(1) & 0xFF==ord('q'):
        break
vid.release()
cv2.destroyAllWindows()
```

```python
vid=cv2.VideoCapture(0) # 0 to open camera 
while True:
    success,frame=vid.read() # to read the video file
    grey=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY) # to convert the video frame to grey scale
    faces=smile_cascade.detectMultiScale(grey,scaleFactor=1.1,minNeighbors=5) # to detect faces in the video frame
    for (x,y,w,h) in faces: # to get the coordinates of the face
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),1) # to draw a rectangle around the face
    cv2.imshow("smile detection",frame) # to display the video frame with the rectangle around the face
    if cv2.waitKey(1) & 0xFF==ord('q'): # to exit the video
        break
vid.release() # to release the video file
cv2.destroyAllWindows() # to destroy all the windows
```
